**EDA of Star Wars Survey Data **

I will perform Exploratory data analysis (EDA) on survey data, surveying 1,186 respondents in America which enabled me to examine what proportions of respondents consideres himserlf or herself as a fan of Star Wars films and which Star Wars films they have watched or not watched

**Objective**

My aim is to explore the dataset with pandas and matplotlib and examine whether and how respondents' gender and age affects the outcome of preference of Star Wars films (whether a fan of the films and which films seen

```python
import pandas as pd
import numpy as np
```

```python
# getting survey CSV data from github source

df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/star-wars-survey/StarWars.csv', encoding = 'unicode_escape')
df.head()
```

Now that I have the survey dataframe ready for analysis, I will do some quick checks and start with looking at two columns first

```python
df.shape
```

There are 1187 rows in the dataframe and the first row is sub-heading which I will be removing shortly, so the dataframe has 1186 (rows) of respondents data for the survey

```python
# removing first row

df = df[1:]
df.head()
```

Now I want to start with looking at two columns and they are

Have you seen any of the 6 films in the Star Wars franchise?

and

Do you consider yourself to be a fan of the Star Wars film franchise?

so I will extract these two columns

```python
cleaned_df = df.iloc[:, 1:3]
cleaned_df
```

```python
cleaned_df.shape
```

Currently the datatype of the two columns is string, I will convert it into boolean for the sake of better data handling using map method




```python
# first, set up a dictionary to turn Yes into True and No into False

yes_no = {
    'Yes': True,
    'No': False
}

# apply .map method to the first column

cleaned_df.iloc[:, 0] = cleaned_df.iloc[:,0].map(yes_no)
cleaned_df.head()
```

```python
# apply .map method to the second column

cleaned_df.iloc[:, 1] = cleaned_df.iloc[:,1].map(yes_no)
cleaned_df.head()
```

Now I can see how many respondents have seen any of the 6 Star Wars films

and how many respondents consider himself/herself to be a Star Wars fan

```python
cleaned_df.apply(pd.value_counts)
```

Quick sanity check on the two columns

```python
cleaned_df.iloc[:,0].value_counts()
```

```python
cleaned_df.iloc[:, 0].isnull().sum()
```

For the first column -

'Have you seen any of the 6 films in the Star Wars franchise?'

all respondents answered either Yes or No

```python
cleaned_df.iloc[:,1].value_counts()
```

```python
cleaned_df.iloc[:, 1].isnull().sum()
```

However for the second column -

'Do you consider yourself to be a fan of the Star Wars film franchise?'

respondents answered Yes, No or null (did not answer the question)

If a survey respondent answered 'Yes' to the second question, meaning the respondent consideres him or her to be a Star Wars film fan, it is very natural to assume he or she must have watched at least one of the six Star Wars films, meaning must have answered the first question -

'Have you seen any of the 6 films in the Star Wars franchise?'

with a 'Yes'

I will do a quick sanity check here to see if any respondents answered as a fan but not seen any of the films

```python
cleaned_df[cleaned_df.iloc[:, 0] == False].apply(pd.value_counts)
```

This tells me that there are NO respondent who answered not watching any Star Wars films but answered as considering himself/herself as a Star Wars fan

I will rename the column names as they are quite long

```python
# renaming column names

columns = ["seen_any_films", "SW_fan"]
cleaned_df.columns = columns

cleaned_df.head()
```

Now I will add two new columns to look at and they are 'Gender' and 'Age'


```python
cleaned_df = cleaned_df.merge(df[['Gender', 'Age']], left_index=True, right_index=True)

cleaned_df.head()
```

Quickly checking new added columns

```python
cleaned_df.iloc[:, -2:].nunique()
```

```python
cleaned_df.iloc[:, -2:].isnull().sum()
```

Both columns have 140 NaN values and 4 types of Age range in the Age column

Now I am going to make a new dataframe based on the current dataframe, showing the number of rows with each unique value or columns so I can work out percentage and make horizontal bar graphs to visualize the overview of the four columns

To do so, I have to also count the number of missing values in each column as it affects the percentage rate as well

```python
# getting the count of unique value of each column including missing values

normalize_df = cleaned_df.apply(lambda x: x.value_counts(dropna=False)).transpose()

normalize_df.head()
```

```python
# now adding normalize parameter to get percentage of each unique value in each columns for horizontal bar plotting

normalize_df = cleaned_df.apply(lambda x: x.value_counts(dropna=False, normalize=True)).transpose()

normalize_df.head()
```

```python
import matplotlib.pyplot as plt

normalize_df.sort_index().plot.barh(   #barh makes it horizontal
    stacked = True,
    rot = 0,   # necessary for x-label title direction
    title = 'Overview of four columns'
    # xlabel = ''
    # xticks = [],  # disabling tick mark in x axis 
)

plt.legend(bbox_to_anchor = (1.05, 1))

plt.show()
```

Now I will take a closer look at horizontal bar chart starting with first two columns



```python
# .sort_values so that True column will be the first column before transpose method is applied

normalize_df2 = cleaned_df.iloc[:, :2].apply(lambda x: x.value_counts(dropna=False)).sort_values('seen_any_films', ascending = False).transpose()

normalize_df2
```

```python
colorlist = ['#ff7f00', 'g', '0.9']

normalize_df2.sort_index().plot.barh(   #barh makes it horizontal
    stacked = True,
    rot = 0,   # necessary for x-label title direction
    title = 'Overview of first two columns',
    edgecolor = 'white',
    color = colorlist
    
    # xlabel = ''
)

plt.legend(bbox_to_anchor = (1.05, 1))

plt.show()
```

```python
normalize_df2_percentage = cleaned_df.iloc[:, :2].apply(lambda x: x.value_counts(dropna=False, normalize=True)).sort_values('seen_any_films', ascending = False).transpose()

normalize_df2_percentage 
```

I can see that ALL 1186 respondents answered the question of the first column, asking if seen any of the 6 Star Wars films

Almost 80% of them have seen the film

For the second column, around 30% of respondents did not answer the question, but almost 47% of the respondents answered that he or she considered himself or herself to be a Star Wars fan


Now I am going to take a look at first column with Gender column to see if whether seen any Star Wars film or not differs by the gender of the respondents

```python
# count value, not normalized

# As .crosstab does not work with dropna = False, will use .fillna instead to include NaN value counts
cross_tab1 = pd.crosstab(index = cleaned_df['seen_any_films'],
                        columns = cleaned_df['Gender'].fillna('Not_answered')
                        # margins=True
                        )
cross_tab1
```

```python
# normalize cross_tab1

cross_tab1_prop= pd.crosstab(index = cleaned_df['seen_any_films'],
                        columns = cleaned_df['Gender'].fillna('Not_answered'), 
                        # margins=True
                        normalize = "index"
                        )

cross_tab1_prop
```

```python
colorlist = ['r', 'b', '0.9']

cross_tab1_prop.plot(kind='barh', 
                        stacked=True, 
                        edgecolor = 'white',
                        color = colorlist,
                        figsize=(10, 6),
                        title = 'Have you seen any of the 6 films in the Star Wars franchise? by Gender'
)


plt.legend(bbox_to_anchor = (1.05, 1))

plt.ylabel("")
plt.xlabel("Proportion")

# count value and proportion percentage text on the bars
for n, x in enumerate([*cross_tab1.index.values]):
    for (proportion, count, y_loc) in zip(cross_tab1_prop.loc[x],
                                          cross_tab1.loc[x],
                                          cross_tab1_prop.loc[x].cumsum()):
                
        plt.text(x=(y_loc - proportion) + (proportion / 2),
                 y=n - 0.11,
                 s=f'{count}\n({np.round(proportion * 100, 1)}%)', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")

plt.show()
```

So what I have done is I have created a cross tab showing the proportion of True and False for whether seen any of six Star Wars films for Gender of the respondents

**True - seen at least one of the six SW films**

Apart from those who did not disclose his/her gender, I would say the gender is about half-half

**False - not seen any of the six SW films**

Unlike proportion for the True, when it comes to respondents who has not seen any Star Wars films, proportion of Female respondents is almost double of Male respondents 

**Finding - Male is more likely to have seen Star Wars films than Female respondents**
